## Exercise: Managing Customer Orders

As part of a business venture, you are starting an online store that sells various products. To ensure smooth operations, you need to develop a program that manages customer orders and inventory.

Follow the steps below to complete the exercise:

1. Define a list called `products` that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".

2. Create an empty dictionary called `inventory`.

3. Ask the user to input the quantity of each product available in the inventory. Use the product names from the `products` list as keys in the `inventory` dictionary and assign the respective quantities as values.

4. Create an empty set called `customer_orders`.

5. Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the `customer_orders` set.

6. Print the products in the `customer_orders` set.

7. Calculate the following order statistics:
   - Total Products Ordered: The total number of products in the `customer_orders` set.
   - Percentage of Products Ordered: The percentage of products ordered compared to the total available products.
   
   Store these statistics in a tuple called `order_status`.

8. Print the order statistics using the following format:
   ```
   Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Percentage of Products Ordered: <percentage_ordered>% 
   ```

9. Update the inventory by subtracting 1 from the quantity of each product. Modify the `inventory` dictionary accordingly.

10. Print the updated inventory, displaying the quantity of each product on separate lines.

Solve the exercise by implementing the steps using the Python concepts of lists, dictionaries, sets, and basic input/output operations. 

````python
#1. list + specific items
products = ["t-shirt", "mug", "hat", "book", "keychain"]
````

````python
#2. dictionary
inventory = {}
#3. quantity
for product in products:
    quantity = input (f"Could you, please, let us know what is the quantity available in stock for {product}?")
    while not quantity.isdigit():
        quantity = input (f"Wrong format. Please insert a valid quantity for {product}!")
    print (f"Available {quantity} {product}!")
    inventory.update({product: int(quantity)})
````

````python
#3. Dictionary verified! 
print (inventory)
````

````python
#4.Empty set
print ("The products selected:")
customer_orders = set()
#5 products wanted
for i in range(3):
    preference = input (f"Could you, please, tell us a product you would like to buy?")
    while not (preference in products):
        preference = input (f"Product not available. Please insert a product in {products}!")
    print (f"You chose: {preference}.")
    customer_orders.add(preference)
#6 Check the set
print("Customer selection:")
print(customer_orders)

#5.1 quantity ordered
print ("The quantities ordered:")
number_orders = {}
for purchased_product in customer_orders:
    quantity_purchased = input(f"How many of {purchased_product} would you like to purchase?")
    while (not quantity_purchased.isdigit()) or (int(quantity_purchased) > inventory[purchased_product]):
        quantity_purchased = input (f"Wrong format or quantity unavailable. The quantity is stock for {purchased_product} is {inventory[purchased_product]}. Please insert a number in {purchased_product}!")
    print (f"You chose: {quantity_purchased} of {purchased_product}.")
    number_orders.update({purchased_product:int(quantity_purchased)})
#5.2 print quantity ordered
print ("Quantity to be purchased:")
print(number_orders)
````

````python
#7-8 Products ordered by the customer
total_products_ordered = sum(number_orders.values())
print (f"The total of products ordered is {total_products_ordered}.")

#7-8 Percentage ordered
total_inventory = sum(inventory.values())
percentage_ordered = (total_products_ordered / total_inventory) *100


#7. Creating a tuple
order_status = (total_products_ordered, percentage_ordered)
print (order_status)
````

````python

#8 Printing the order status:

print(f"Order Statistics: \n Total Products Ordered: {order_status[0]}\n Percentage of Products Ordered: {order_status[1]} %.")
````

````python
#9 New inventory
for product, quantity in inventory.items():
    inventory.update({product: quantity-1})
````

````python
#10 Printing the update inventory
for product, quantity in inventory.items():
    print (f"The quantity of {product} is {quantity}.\n")
````
